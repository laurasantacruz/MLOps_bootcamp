```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the datasets into dataframes
circuits = pd.read_csv("f1_files/circuits.csv", low_memory=False)
constructor_results = pd.read_csv("f1_files/constructor_results.csv",low_memory=False)
constructor_standings = pd.read_csv("f1_files/constructor_standings.csv",low_memory=False)
constructors = pd.read_csv("f1_files/constructors.csv",low_memory=False)
driver_standings = pd.read_csv("f1_files/driver_standings.csv",low_memory=False)
drivers = pd.read_csv("f1_files/drivers.csv",low_memory=False)
lap_times = pd.read_csv("f1_files/lap_times.csv",low_memory=False)
pit_stops = pd.read_csv("f1_files/pit_stops.csv",low_memory=False)
qualifying = pd.read_csv("f1_files/qualifying.csv",low_memory=False)
races = pd.read_csv("f1_files/races.csv",low_memory=False)
results = pd.read_csv("f1_files/results.csv",low_memory=False)
seasons = pd.read_csv("f1_files/seasons.csv",low_memory=False)
sprint_results = pd.read_csv("f1_files/sprint_results.csv",low_memory=False)
status = pd.read_csv("f1_files/status.csv",low_memory=False)
```

```python
# Merge in a single dataframe
# the results file will be my base table 
join_1 = results.merge(races, how='left', on = 'raceId')
join_1.rename(columns = {'time_x':'time_results','time_y':'time_races'}, inplace = True)

join_2 = join_1.merge(drivers, how='left', on='driverId')
join_2.rename(columns = {'number_x':'number_results','url_x':'url_races','number_y':'number_drivers','url_y':'url_drivers'}, inplace=True)

join_3 = join_2.merge(constructors, how='left', on='constructorId')
join_3.rename(columns={'name_x':'name_race','name_y':'constructor_name','nationality_x':'driver_nationality','nationality_y':'constructor_nationality'},inplace=True)

join_4 = join_3.merge(status, how='left', on='statusId')


```

```python
final_df = join_4[['grid','position','points','laps','time_results','milliseconds','fastestLap',
                  'fastestLapTime','fastestLapSpeed','year','name_race','date','forename','surname',
                  'driver_nationality','constructor_name','constructor_nationality','status']]
```

```python
### Change data types
cols = ['milliseconds', 'fastestLap', 'fastestLapSpeed']
final_df[cols] = final_df[cols].apply(pd.to_numeric, errors='coerce', axis=1)
final_df.dtypes
```

```python
# Since we have a lot of data I will only focus on the last 5 years
final_df2 = final_df[final_df.year > 2017]

dfm = final_df2.groupby(['year','constructor_name']).sum()
dfm_2 = dfm[['points']]
pd.set_option('display.max_rows', dfm_2.shape[0]+1)
print(dfm_2)
```

```python
fastest_lap_avg = final_df2.groupby('constructor_name')['fastestLapSpeed'].mean()
fastest_lap_avg.head()
```

```python
for constructor in final_df2['constructor_name'].unique():
    constructor_to_plot = final_df2[final_df2['constructor_name'] == constructor]
    fig = plt.figure(figsize = (10, 5))
 
    # creating the bar plot
    plt.bar(constructor_to_plot['year'], constructor_to_plot['points'], color ='maroon',
            width = 0.4)

    plt.xlabel("Year")
    plt.ylabel("No. of points")
    plt.title(f"{constructor}: Points by year")
    plt.show()
```
